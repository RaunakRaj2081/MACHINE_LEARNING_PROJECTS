# Diabetes Prediction using Support Vector Machine (SVM)


 1. Introduction to Diabetes Prediction

=> Diabetes prediction is a classification problem in which we try to predict whether a person has diabetes or not, based on several medical input features such as:

Glucose level

Blood pressure

BMI (Body Mass Index)

Age, Insulin, etc.

We use machine learning techniques like Support Vector Machine (SVM) to build an accurate prediction model.

 2. What is Support Vector Machine (SVM)?

=> Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. In diabetes prediction, SVM classifies patients into two classes:

1 (diabetic)

0 (non-diabetic)

SVM tries to find the best decision boundary (hyperplane) that separates these two classes with maximum margin.

 3. Dataset Used

=> The most common dataset used for diabetes prediction is the PIMA Indian Diabetes Dataset, which includes:

768 instances

8 input features (Glucose, BloodPressure, BMI, etc.)

1 output label (Outcome: 1 for diabetic, 0 for non-diabetic)

 4. Steps Involved in Diabetes Prediction using SVM

a. Data Collection:

Use a dataset containing medical information of patients.

b. Data Preprocessing:

Handle missing values, clean the data, and standardize it using scaling techniques.

c. Train-Test Split:

Divide the dataset into training and testing sets using train_test_split().

d. Feature Scaling:

Use StandardScaler() to normalize the data, since SVM is sensitive to the scale of features.

e. Model Training:

Train the SVM classifier using the training data with a specific kernel (e.g., linear, rbf).

f. Prediction:

Make predictions using model.predict() on new or test data.

g. Evaluation:

Check model performance using metrics like accuracy score, confusion matrix, and classification report.

 5. Kernel Trick in SVM

=> SVM can work with non-linearly separable data by using the kernel trick. Some common kernels are:

Linear Kernel: For linearly separable data

RBF Kernel (Radial Basis Function): For non-linear data

Polynomial Kernel: For polynomial decision boundaries



#Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

#Data Collection and Analysis

PIMA Diabetes Dataset

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
from re import X
#separating the data and the labels
X = diabetes_dataset.drop(columns='Outcome' , axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

#Data Standardization

```python
scaler = StandardScaler()
```

```python
standardized_data = scaler.fit_transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

#Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

#Model Evaluation

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

#Making a Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
